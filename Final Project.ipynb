```python
# Packages for data manipulation 
import pandas as pd
import numpy as np

# Packages for ploting
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

# Package to filter warnings
import warnings 
warnings.filterwarnings('ignore')

# Packages for ML
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn import preprocessing
from sklearn.decomposition import PCA
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
# FE
# Get rid of unnecessary categories
```

```python
# Education category 0 5 6 can be combined in category 4

df['EDUCATION']=np.where(df['EDUCATION'] == 5, 4, df['EDUCATION'])
df['EDUCATION']=np.where(df['EDUCATION'] == 6, 4, df['EDUCATION'])
df['EDUCATION']=np.where(df['EDUCATION'] == 0, 4, df['EDUCATION'])
df.EDUCATION.unique()
```

```python
# Marriage category 0 can be combined in other as we do not have any information regarding it

df['MARRIAGE']=np.where(df['MARRIAGE'] == 0, 3, df['MARRIAGE'])
df.MARRIAGE.unique()
```

```python
# rename target column to default for easy access 

df = df.rename(columns={'default.payment.next.month':'DEFAULT'})
```

```python
# Logistic regression using all independent variable
```

```python
X = df[['LIMIT_BAL', 'SEX', 'EDUCATION', 'MARRIAGE', 'AGE', 
        'PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6', 
        'BILL_AMT1','BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6',
        'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']]
y = df[['DEFAULT']]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.3,random_state=30,shuffle=True)
```

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
lr = LogisticRegression()
lr.fit(X_train,y_train)

y_pred_lr = lr.predict(X_test)
```

```python
print('Accuracy:',accuracy_score(y_pred_lr,y_test)*100,'%')
print('--------------------------------------------------------------------------------------------------------')
print(classification_report(y_test,y_pred_lr))
print('--------------------------------------------------------------------------------------------------------')
cnf_lr = confusion_matrix(y_test,y_pred_lr)
sns.heatmap(cnf_lr,annot=True, cmap="rocket", fmt="d", xticklabels = ['No default', 'Default'], yticklabels = ['No default', 'Default'])
plt.title('Confusion matrix heatmap')
plt.xlabel('Predicted')
plt.ylabel('Actual')
```
